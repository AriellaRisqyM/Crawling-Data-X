```python
import os
import zipfile
from google.colab import files

# Nama folder yang ingin di-zip
folder_to_zip = 'tweets-data' #ganti sesuai nama folder hasil crawling

# Nama file zip yang akan dihasilkan
#ganti sesuai keinginan untuk nama file zip jangan lupa .zip
zip_filename = 'data_archive.zip' 

print(f"ðŸ“¦ Mengompres folder '{folder_to_zip}'...")

# Membuat file zip
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    # Berjalan melalui semua file di dalam folder
    for root, dirs, files_in_dir in os.walk(folder_to_zip):
        for filename in files_in_dir:
            # Membuat path lengkap ke file
            file_path = os.path.join(root, filename)
            # Menambahkan file ke dalam zip
            zipf.write(file_path, os.path.relpath(file_path, os.path.join(folder_to_zip, '..')))

print(f"âœ… Berhasil membuat file '{zip_filename}'.")
print("ðŸš€ Memulai proses unduh...")

# Memicu unduhan
files.download(zip_filename)
```

```python
import os
import pandas as pd
import glob
from datetime import date, timedelta

# 1. auth token Twitter API
twitter_auth = 'oqneeenf90223fcasxxxxxxxy'

# 2. Keyword pencarian
keyword = "" #contoh: "(vaksin OR vaksinasi OR vaksinasi covid19)(corona OR covid19 OR covid-19)"

# variabel tanggal mulai dan akhir, serta limit harian
#format tanggal: 'YYYY-MM-DD'
start_date = '2024-01-01'
end_date = '2024-12-30'
Limit_day = 50 #jumlah tweet per hari bukan perbulan jadi akan melakukan crawling data per hari/tanggal

# Proses crawling
start_date = date.fromisoformat(start_date)
end_date = date.fromisoformat(end_date)
current_date = start_date

while current_date <= end_date:
    date_str = current_date.strftime('%Y-%m-%d')
    print(f"\nCrawling untuk tanggal: {date_str}...")

    since_str = date_str
    until_str = (current_date + timedelta(days=1)).strftime('%Y-%m-%d')

    filename = f'2024_data_harian_{date_str}.csv' #nama file hasil crawling

    search_keyword = f"{keyword} lang:id since:{since_str} until:{until_str}" #lang:id untuk bahasa Indonesia, lang:en untuk bahasa Inggris
    # Perintah untuk menjalankan tweet-harvest
    command = f'npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {Limit_day} --token {twitter_auth}'

    os.system(command)

    current_date += timedelta(days=1)
```
